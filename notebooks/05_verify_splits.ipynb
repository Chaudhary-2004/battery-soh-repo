```python
# ================================
# VERIFY YOUR SPLIT FILES
# ================================

import pandas as pd
import numpy as np
import os

print("🔍 VERIFYING YOUR SPLIT FILES")
print("=" * 50)

splits_dir = '../DATA/splits'

# Check if all files exist
required_files = ['X_train.csv', 'X_val.csv', 'X_test.csv', 
                  'y_train.csv', 'y_val.csv', 'y_test.csv']

missing_files = []
for file in required_files:
    filepath = os.path.join(splits_dir, file)
    if os.path.exists(filepath):
        size_kb = os.path.getsize(filepath) / 1024
        print(f"✅ {file}: {size_kb:.1f} KB")
    else:
        missing_files.append(file)
        print(f"❌ {file}: MISSING!")

if missing_files:
    print(f"\n⚠️  Missing files: {missing_files}")
    exit()

# Load and check the splits
print("\n📊 LOADING AND CHECKING SPLITS...")
print("-" * 40)

# Load feature matrices
X_train = pd.read_csv(f'{splits_dir}/X_train.csv')
X_val = pd.read_csv(f'{splits_dir}/X_val.csv')
X_test = pd.read_csv(f'{splits_dir}/X_test.csv')

# Load target vectors
y_train = pd.read_csv(f'{splits_dir}/y_train.csv')
y_val = pd.read_csv(f'{splits_dir}/y_val.csv')
y_test = pd.read_csv(f'{splits_dir}/y_test.csv')

print(f"📈 Train Set: X{X_train.shape}, y{y_train.shape}")
print(f"📊 Val Set:   X{X_val.shape}, y{y_val.shape}")
print(f"📉 Test Set:  X{X_test.shape}, y{y_test.shape}")

# Check feature consistency
print(f"\n🔧 Features: {X_train.shape[1]} columns")
print(f"   {list(X_train.columns)}")

# Check for missing values
print(f"\n🔍 Missing Values Check:")
print(f"   X_train: {X_train.isnull().sum().sum()} missing")
print(f"   X_val:   {X_val.isnull().sum().sum()} missing")
print(f"   X_test:  {X_test.isnull().sum().sum()} missing")

# Check target distribution
print(f"\n🎯 Target (SOH) Statistics:")
print(f"   Train - Mean: {y_train.iloc[:,0].mean():.3f}, Std: {y_train.iloc[:,0].std():.3f}")
print(f"   Val   - Mean: {y_val.iloc[:,0].mean():.3f}, Std: {y_val.iloc[:,0].std():.3f}")
print(f"   Test  - Mean: {y_test.iloc[:,0].mean():.3f}, Std: {y_test.iloc[:,0].std():.3f}")

print("\n✅ SPLITS VERIFICATION COMPLETE!")
print("🚀 Ready for machine learning!")
```

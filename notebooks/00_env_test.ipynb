```python
import os
import random
import numpy as np
import pandas as pd
import sklearn
import xgboost
import matplotlib.pyplot as plt

# -----------------------
# 5.5 Reproducibility
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)

# -----------------------
# Print package versions
print('numpy', np.__version__)
print('pandas', pd.__version__)
print('sklearn', sklearn.__version__)
print('xgboost', xgboost.__version__)

# -----------------------
# Quick load to test
# df = pd.read_csv('data/processed/all_cells_cycle_features.csv', nrows=10)
# print(df.head())

# -----------------------
# Small sanity plot
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title('Sanity Plot: sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
